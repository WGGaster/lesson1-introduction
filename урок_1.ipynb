```python
#2.1
x = t.tensor(6., requires_grad=True)
y = t.tensor(3., requires_grad=True)
z = t.tensor(9., requires_grad=True)

f = x ** 2 + y ** 2 + z ** 2 + 2 * x * y * z
print(f)
print()
f.backward()

print('градиенты:')
print(x.grad)
print(y.grad)
print(z.grad)

print()

x = x.item()
y = y.item()
z = z.item()

print('проверка аналитически:')
print(2 * x + 2 * y * z)
print(2 * y + 2 * x * z)
print(2 * z + 2 * y * x)
```

```python
#2.2
def mse(y_pred, y_true):
  return t.mean((y_pred - y_true) ** 2)

t.manual_seed(42)
n = 5
x = t.randn(5)
y_true = 2 * x + 3.4 + t.randn(n)
w = t.tensor(2., requires_grad=True)
b = t.tensor(0., requires_grad=True)
y_pred = w * x + b
print(loss := mse(y_pred, y_true))
loss.backward()
print(w.grad, b.grad)
```

```python
#2.3
def sin(x):
  return t.sin(x ** 2 + 1)


x1 = t.tensor(2., requires_grad=True)
y1 = sin(x1)
y1.backward()
print(x1.grad)

x2 = t.tensor(2., requires_grad=True)
y2 = sin(x2)
print(t.autograd.grad(y2, x2)[0])
```

```python
#3.1
t.manual_seed(42)

a = t.rand(64, 1024, 1024)
b = t.rand(128, 512, 512)
c = t.rand(256, 256, 256)
```

```python
#3.2

def measure_time_cpu(func, *args, **kwargs):
    start_time = time.time()
    _ = func(*args, **kwargs)
    end_time = time.time()
    return end_time - start_time

def measure_time_gpu(func, *args, **kwargs):
    start_event = t.cuda.Event(enable_timing=True)
    end_event = t.cuda.Event(enable_timing=True)

    start_event.record()
    _ = func(*args, **kwargs)
    end_event.record()

    end_event.synchronize()
    elapsed = start_event.elapsed_time(end_event)
    return elapsed
```

```python
#3.3
import pandas as pd

device = 'cuda' if t.cuda.is_available() else 'cpu'
a_gpu = a.to(device)
b_gpu = b.to(device)
c_gpu = c.to(device)

# Инициализация списка для хранения результатов
results = []

# Определение операций для сравнения (используем совместимые тензоры)
operations = [
    ("Матричное умножение", t.matmul, (a_gpu, a_gpu.transpose(-1, -2))),
    ("Поэлементное сложение", t.add, (a_gpu, a_gpu)),
    ("Поэлементное уумножение", t.mul, (a_gpu, a_gpu)),
    ("Транспонирование", t.transpose, (a_gpu, 0, 1)),
    ("Сумма всех элементов", t.sum, (a_gpu,))
]

# Измерение времени для каждой операции (только один запуск измерений)
for op_name, op_func, op_args in operations:
    cpu_ms = measure_time_cpu(op_func, *op_args)
    if device == 'cuda':
        gpu_ms = measure_time_gpu(op_func, *op_args)
        speedup = cpu_ms / gpu_ms if gpu_ms and gpu_ms > 0 else None
    else:
        gpu_ms = None
        speedup = None

    results.append({
        "Операция": op_name,
        "CPU мс": cpu_ms,
        "GPU мс": gpu_ms if gpu_ms is not None else "N/A",
        "Ускорение": f"{speedup:.2f}x" if speedup is not None else "N/A"
    })

# Показать результаты
results_df = pd.DataFrame(results)
print(results_df)
```
